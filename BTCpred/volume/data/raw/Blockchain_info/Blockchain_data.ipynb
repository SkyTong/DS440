# Hourly total exchange vol

```python
import requests
import datetime

# Function to fetch data
def fetch_data(fsym, tsym, to_ts):
    api_key = 'e4183aca8d21b1e16a8be6348e27b8b36264fd0493c7f42f89ecabf5c391fd09'
    base_url = "https://min-api.cryptocompare.com/data/v2/histohour"
    params = {
        'fsym': fsym,
        'tsym': tsym,
        'limit': 2000,
        'toTs': to_ts,
        'api_key': api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        return response.json()['Data']['Data']
    else:
        raise Exception(f"API request failed with status code {response.status_code}")

# Initial setup
fsym = 'BTC'
tsym = 'USD'
end_date = datetime.datetime.now()
start_date = datetime.datetime(2022, 2, 10)
all_data = []

# Convert end_date to timestamp
current_ts = end_date.timestamp()

# Loop to fetch data iteratively
while True:
    data = fetch_data(fsym, tsym, current_ts)
    if not data:
        break  # Break the loop if no data is returned
    earliest_ts = data[0]['time']  # Get the earliest timestamp from the fetched data
    all_data = data + all_data  # Concatenate in a way that new data is added at the beginning
    if earliest_ts <= start_date.timestamp():
        break  # If the earliest data is before or on the start date, stop fetching more data
    current_ts = earliest_ts  # Update current_ts to fetch the next chunk


# Reverse to maintain chronological order
#all_data.reverse()

# Optionally, process all_data as needed
# For example, print out the first and last data points to confirm
if all_data:
    first_point = all_data[0]
    last_point = all_data[-1]
    print(f"First data point: Time: {datetime.datetime.fromtimestamp(first_point['time'])}, Volume: {first_point['volumeto']}")
    print(f"Last data point: Time: {datetime.datetime.fromtimestamp(last_point['time'])}, Volume: {last_point['volumeto']}")
else:
    print("No data fetched")
```

# Hourly OHLCV


```python
import requests
import datetime

def fetch_histohour_data(fsym, tsym, to_ts):
    api_key = 'your_api_key'  # Replace with your actual API key
    base_url = "https://min-api.cryptocompare.com/data/v2/histohour"
    params = {
        'fsym': fsym,  # From Symbol, e.g., 'BTC'
        'tsym': tsym,  # To Symbol, e.g., 'USD'
        'limit': 2000,  # Maximum number of data points to fetch
        'toTs': to_ts,  # End timestamp for fetching data
        'api_key': api_key
    }
    response = requests.get(base_url, params=params)
    if response.status_code == 200:
        return response.json()['Data']['Data']
    else:
        raise Exception(f"Failed to fetch data: {response.status_code}")

# Set your parameters here
fsym = 'BTC'
tsym = 'USD'
end_date = datetime.datetime.now()  # End date for the data
start_date = datetime.datetime(2022, 2, 10)  # Start date for the data
all_data = []

# Convert end_date to a timestamp
current_ts = end_date.timestamp()

# Fetch data in chunks, adjusting `toTs` to cover the period from start_date to end_date
while True:
    data = fetch_histohour_data(fsym, tsym, current_ts)
    if not data:
        break  # Break the loop if no data is returned
    earliest_ts = data[0]['time']  # Get the earliest timestamp from the fetched data
    all_data = data + all_data  # Prepend the fetched data
    if earliest_ts <= start_date.timestamp():
        break  # If the earliest data is before or on the start date, stop fetching more data
    current_ts = earliest_ts  # Update current_ts to fetch the next chunk


# Optionally, print the first and last data points to verify
if all_data:
    first_point = all_data[0]
    last_point = all_data[-1]
    print(f"First data point: Time: {datetime.datetime.fromtimestamp(first_point['time'])}, Price: {first_point['close']}")
    print(f"Last data point: Time: {datetime.datetime.fromtimestamp(last_point['time'])}, Price: {last_point['close']}")
else:
    print("No data fetched")

# Specify CSV file name
csv_file_name = 'hourly_OHLCV.csv'

# Write data to CSV
with open(csv_file_name, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['time', 'high', 'low', 'open', 'close', 'volumefrom', 'volumeto'])
    for item in all_data:
        writer.writerow([datetime.datetime.fromtimestamp(item['time']).strftime('%Y-%m-%d %H:%M:%S'),
                         item['high'], item['low'], item['open'], item['close'], item['volumefrom'], item['volumeto']])

csv_file_name
```
